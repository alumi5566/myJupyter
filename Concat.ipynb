```python
import pandas as pd
```

```python
df1=pd.DataFrame({"A":["A0","A1","A2","A3"],"B":["B0","B1","B2","B3"],\
                  "C":["C0","C1","C2","C3"],"D":["D0","D1","D2","D3"],\
                      "E":["E0","E1","E2","E3"]})
df1
```

```python
df2=pd.DataFrame({"A":["A4","A5","A6","A7"],"B":["B4","B5","B6","B7"],\
                  "C":["C4","C5","C6","C7"],"D":["D4","D5","D6","D7"],\
                      "F":["F4","F5","F6","F7"]})
df2
```

```python
# 預設 axis=0 合併row, join=outer , ignore_index=False
result=pd.concat([df1,df2])
result
```

```python
result2 = pd.concat([df1,df2], join="inner")
result2
```

```python
result3 = pd.concat([df1,df2], axis=1, join="outer")
result3
```
